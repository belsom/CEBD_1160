```python
# Question 5. Drop rows with null values
train.dropna() #only one age column contains null values
#train.info()

## Dropping all columns with null values
#train.dropna(axis=1)
```

```python
# Question 6. Plot fare histogram by index 
plt.hist(train['Fare'], bins=3, color='g')
plt.title('Fare by Index')
plt.xlabel('Index')
plt.ylabel('Fare')
```

```python
# Question 6. Plotting fare histogram
train.hist(column='Fare', figsize=(12,9), bins=5)         
plt.savefig(f'plots/fare_hist.png', format='png')
```

```python
# Question 7. Plotting Survived histogram
train.hist(column='Survived', figsize=(12,9), bins=5)         
plt.savefig(f'plots/survived_hist.png', format='png')
```

```python
#Plot survived histogram
#plt.figure(figsize=(8, 6))
#sns.distplot(train['Survived'], color='g', bins=100, hist_kws={'alpha': 0.4});
plt.hist(train['Survived'], bins=3, color='g')
plt.title('Survived')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.savefig(f'plots/survived_hist_count.png', format='png')
#plt.clf()
```

```python
# Question 8. Draw correlation matrix
fig, ax = plt.subplots(figsize=(12,12))
sns.heatmap(train.corr(), annot=True, cmap='autumn')
ax.set_xticklabels(train.columns, rotation=45)
ax.set_yticklabels(train.columns, rotation=45)
```

```python
corr = train.corr() 
plt.figure(figsize=(12, 10))

sns.heatmap(corr[(corr >= -1)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True);
```

```python
# Advanced 
# Question 1 :Create a new column called “doubledFare” and double the fares. Append the new column to your dataframe
train.insert(10,'doubledFare', train['Fare'] * 2)
train.info()
```

```python
# Question 2. Knowing the value of mean to answer the question
train['Age'].mean()
```

```python
# Question 2. Fill null age values by mean(age)
train['Age'] = train['Age'].fillna((train['Age'].mean()))
#train
```

```python
# Question 3. Scatter plot age to fare and interpret
#scatter(train.Age, train.Fare) 
plt.scatter(train['Age'], train['Fare'], color='b')
plt.title('Age to Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.savefig(f'plots/age_to_fare.png', format='png')
```

```python
#Present using groupby the count, mean, and standard deviation of survived column
SurvivedCount = train.groupby(["Survived"],as_index = False).count()
SurvivedMean = train.groupby(["Survived"],as_index = False).mean()
SurvivedStd = train.groupby(["Survived"],as_index = False).std()
```
